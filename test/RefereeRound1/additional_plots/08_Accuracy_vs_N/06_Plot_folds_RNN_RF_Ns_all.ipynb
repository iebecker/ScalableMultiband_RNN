```python
import pandas as pd
import numpy as np
from glob import glob
import matplotlib.pyplot as plt
from sklearn.metrics  import classification_report
```

Plot the accuracy vs the number of observations for our RNN.

```python
# survey = 'Gaia'
survey='PanStarrs'
# survey = 'ZTF'

# surveys = ['Gaia', 'PanStarrs', 'ZTF']
surveys = ['Gaia', 'PanStarrs2', 'ZTF']
```

```python
plt.rcParams['hatch.linewidth'] = 0.5
```

```python
fmt = {'fontsize':18, 'fontname':'Times New Roman'}
fmt_title = {'fontsize':22, 'fontname':'Times New Roman'}

    
fig, axs = plt.subplots(3, figsize=(20,13))
fig.suptitle('Average macro F-score as a function of the total number of observations', **fmt_title)

plt.subplots_adjust(
                    # left=0.1,
                    # bottom=0.1,
                    # righ/t=0.9,
                    # top=0.9,
                    # wspace=0.4,
                    hspace=0.3
                    )

for ii in range(len(surveys)):
    survey = surveys[ii]
    # Load the metadata, to get the length.
    if survey=='Gaia':
        path_metadata = './../../../../../Prepare_dataset/Gaia/V5/Dataset_Gaia_Phys_V5.dat'
        cols = ['ID','N', 'N_b', 'N_r']
    elif survey=='PanStarrs':
        path_metadata = '../../../../../Prepare_dataset/PanStarrs/Dataset_PanStarrs_V2_ParamPhys.dat'
        cols =['ID','N','N_g', 'N_r', 'N_i', 'N_z', 'N_y']
    elif survey=='PanStarrs2':
        path_metadata = '../../../../../Prepare_dataset/PanStarrs/Dataset_PanStarrs_V2_ParamPhys.dat'
        cols =['ID','N','N_g', 'N_r', 'N_i', 'N_z', 'N_y']        
    elif survey=='ZTF':
        path_metadata = '../../../../../../Databases/ALeRCE/Snchez-Sez et al.2021/metadata.csv'
        cols =['ID', 'N','N_1', 'N_2']


    # Load the metadata
    metadata = pd.read_csv(path_metadata, usecols=cols)

    path = '../04_Compute_Metrics/'+survey+'/Results/*.dat'
    folds_rnn = glob(path)
    folds_rnn.sort()

    path_rf = '../../../../../RandomForestBaseline/Compute_TurboFATS_v2/'+survey+'/Results/*.dat'
    folds_rf = glob(path_rf)
    folds_rf.sort()

    dfs_rnn = []
    for fold in folds_rnn:

        # Load RNN results
        pred_rnn = pd.read_csv(fold)
        pred_rnn.index.name = 'ID'

        df_rnn = pd.merge(pred_rnn[['Class', 'Pred']], metadata[cols], on='ID', how='inner')    
        dfs_rnn.append(df_rnn)
    dfs_rf = []
    for fold in folds_rf:
       # Load RF results
        pred_rf = pd.read_csv(fold)
        pred_rf.rename({'oid':'ID'}, axis=1, inplace=True)
        pred_rf.set_index('ID', inplace=True)
        df_rf = pd.merge(pred_rf[['Class', 'Pred']], metadata[cols], on='ID', how='inner')        
        dfs_rf.append(df_rf)

    # Set the default interval each 10 observations, using the first fold
    df_rnn = dfs_rnn[0]
    df_rf = dfs_rf[0]

    if survey=='Gaia':
        interval = 10
        delta = interval/2
        N_max = np.ceil(df_rnn.N.max()/interval)
        XX = np.arange(25, N_max*interval, interval)
    if survey=='PanStarrs':
        interval = 10
        delta = interval/2
        N_max = np.ceil(df_rnn.N.max()/interval)
        XX = np.arange(10, N_max*interval, interval)
    if survey=='PanStarrs2':
        interval = 10
        delta = interval/2
        N_max = np.ceil(df_rnn.N.max()/interval)
        XX = np.arange(10, N_max*interval, interval)        
    if survey=='ZTF':
        interval = 50
        delta = interval/2
        N_max = np.ceil(df_rnn.N.max()/interval)
        XX = np.arange(10, N_max*interval, interval)     

    # Compute the number of elements per bin using default values
    N = []
    deltas = []
    for i in range(len(XX)):
        x = XX[i]
        deltas.append(delta)
        # Define the intervals
        below = x-delta
        above = x+delta
        # Create the bined dataframe for each model type
        b1 = df_rnn.N>=below
        b2 = df_rnn.N<above
        b = np.logical_and(b1, b2)
        sel_rnn = df_rnn[['Class', 'Pred', 'N']][b]

        N.append(sel_rnn.shape[0])

    # Some bins have a vero low count. Merge bins of the boundaries.
    new_XX = list(XX).copy()
    new_N = N.copy()
    new_deltas = deltas.copy()

    N_min = 10
    condition_while=True
    new_start = 0

    while_exit=0
    while condition_while:
        cond_for = False

        for i in range(new_start, len(new_N)):
            N_temp = new_N[i]
            X_temp = new_XX[i]
            d_temp = new_deltas[i]
            if N_temp<N_min and i<len(new_N)-1:
                # Create the new bin
                X_temp = (new_XX[i]+new_XX[i+1])/2
                d_temp = new_deltas[i]+deltas[i+1]
                N_temp = new_N[i]+new_N[i+1]

                # Remove the last one
                new_XX.pop(i+1)
                new_deltas.pop(i+1)
                new_N.pop(i+1)

                # Replace the new values
                new_XX[i] = X_temp
                new_deltas[i] = d_temp
                new_N[i] = N_temp

                # Break the for loop and start again
                cond_for = True

            elif N_temp<N_min and i==len(new_N)-1:
                # Create the new bin
                X_temp = (new_XX[i-1]+new_XX[i])/2
                d_temp = new_deltas[i-1]+deltas[i]
                N_temp = new_N[i-1]+new_N[i]            

                # Remove the last one
                new_XX.pop(i)
                new_deltas.pop(i)
                new_N.pop(i)

                # Replace the new values
                new_XX[i-1] = X_temp
                new_deltas[i-1] = d_temp
                new_N[i-1] = N_temp

                condition_while=False   

            # If found a bin, break
            if cond_for:
                break    
        while_exit+=1
        if while_exit>25:
            condition_while=False
    new_XX = np.array(new_XX).astype(int)
    new_deltas = np.array(new_deltas)        

    for i in range(len(new_XX)):

        REPORT = {'RNN':[], 'RF':[]}
        ACC = {'RNN':[], 'RF':[]}    
        # Define the intervals
        below = new_XX[i]-new_deltas[i]
        above = new_XX[i]+new_deltas[i]    
        # For each fold
        for j in range(len(dfs_rnn)):
            # Get the corresponding results of RNN
            df_rnn = dfs_rnn[j]
            # Create the bined dataframe for each model type
            b1 = df_rnn.N>=below
            b2 = df_rnn.N<above
            b = np.logical_and(b1, b2)
            sel_rnn = df_rnn[['Class', 'Pred']+cols[1:]][b]
        Ns = {c[-1]:np.round(sel_rnn[c].mean()/sel_rnn[cols[1]].mean(), 2) for c in cols[2:]}

    # Create the containers
    REPORTs = []
    ACCs = []

    # For each bin
    for i in range(len(new_XX)):

        REPORT = {'RNN':[], 'RF':[]}
        ACC = {'RNN':[], 'RF':[]}    
        # Define the intervals
        below = new_XX[i]-new_deltas[i]
        above = new_XX[i]+new_deltas[i]    
        # For each fold
        for j in range(len(dfs_rnn)):
            # Get the corresponding results of RNN
            df_rnn = dfs_rnn[j]
            # Create the bined dataframe for each model type
            b1 = df_rnn.N>=below
            b2 = df_rnn.N<above
            b = np.logical_and(b1, b2)
            sel_rnn = df_rnn[['Class', 'Pred', 'N']][b]

            # Get the corresponding results of RF
            df_rf = dfs_rf[j]
            # Create the bined dataframe for each model type
            b1 = df_rf.N>=below
            b2 = df_rf.N<above
            b = np.logical_and(b1, b2)
            sel_rf = df_rf[['Class', 'Pred', 'N']][b]

            # Compute the reports
            rep_RNN = classification_report(sel_rnn['Class'],sel_rnn['Pred'], output_dict=True, zero_division=0)
            rep_RF = classification_report(sel_rf['Class'],sel_rf['Pred'], output_dict=True, zero_division=0)

            # Aggregate results

            REPORT['RNN'].append(rep_RNN)
            REPORT['RF'].append(rep_RF)
        REPORTs.append(REPORT)


    F1 = {'RNN':[], 'RF':[]}

    for key in ['RNN', 'RF']:
        for bin_ in range(len(REPORTs)):

            report = REPORTs[bin_][key]
            F1s = [report[i]['macro avg']['f1-score'] for i in range(len(report))]

            mean, std = np.mean(F1s), np.std(F1s)
            F1[key].append([mean, std])
        F1[key] = np.array(F1[key])



    # Custom ticks 
    if survey=='Gaia':
        xticks = [i for i in range(new_XX.shape[0])  if new_XX[i]>125 or i%2==0  ]
        delta = 2
        width_ = interval-6
    elif survey=='PanStarrs':
        xticks = [i for i in range(new_XX.shape[0])]
        delta = 2
        width_ = interval-6
    elif survey=='PanStarrs2':
        xticks = [i for i in range(new_XX.shape[0])]
        delta = 2
        width_ = interval-6        
    elif survey=='ZTF':
        xticks = [i for i in range(new_XX.shape[0])]
        width_ = interval/3
        delta = width_/2    

    axs[ii].bar(new_XX-delta, F1['RNN'][:,0], width=width_, label='RNN', hatch="--")
    axs[ii].errorbar(new_XX-delta, F1['RNN'][:,0], yerr=F1['RNN'][:,1],fmt='none', color='black', elinewidth=2)

    axs[ii].bar(new_XX+delta, F1['RF'][:,0], width=width_, label='BRF', hatch="//")
    axs[ii].errorbar(new_XX+delta, F1['RF'][:,0], yerr=F1['RF'][:,1],fmt='none', color='black', elinewidth=2)

    axs[ii].set_xticks(new_XX[xticks])
    if survey=='PanStarrs2':
        survey='Pan-STARRS1'
    axs[ii].set_title(survey, **fmt)
    # Put a legend below current axis
axs[0].legend(loc='upper right',
              bbox_to_anchor=(1.0, 1.2),
              fontsize=16,
              fancybox=True, shadow=False, ncol=2)    
axs[0].set_ylabel('F-score', **fmt)
axs[1].set_ylabel('F-score', **fmt)
axs[2].set_ylabel('F-score', **fmt)

axs[0].set_ylim(0,1.1)
axs[1].set_ylim(0,1.1)
axs[2].set_ylim(0,1.1)

axs[2].set_xlabel('Total number of observations', **fmt)
plt.savefig('surveys.png', dpi=250,format='png', bbox_inches='tight' )
plt.savefig('surveys.pdf', dpi=250,format='pdf', bbox_inches='tight' )
plt.show()    
```
